```python
import joblib
import os
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
 
import matplotlib.pyplot as plt
import seaborn as sns
 
from scipy.stats import randint
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, label_binarize
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.feature_selection import VarianceThreshold
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix,
    classification_report, roc_curve, auc
)
 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
 
print('All libraries imported successfully.')
```

```python
FILE_PATH = '/kaggle/input/datasets/aaabdul/bccc-cpacket-cloud-ddos-2024-dataset/merged_CSVs.csv'
print('Loading dataset...')
fmt = pd.read_csv(FILE_PATH, low_memory=False)
fmt.columns = fmt.columns.str.strip()
print(f'Loaded: {fmt.shape[0]:,} rows × {fmt.shape[1]} columns')
fmt.head(5)
```

```python
LABEL_COL = 'label'
print("Class distribution")
print(fmt[LABEL_COL].value_counts())
 
plt.figure(figsize=(8, 4))
fmt[LABEL_COL].value_counts().plot(kind='bar', color=['steelblue', 'red', 'yellow'])
plt.title('Class Distribution')
plt.xlabel('Label')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
before = len(fmt)
fmt.drop_duplicates(inplace=True)
fmt.reset_index(drop=True, inplace=True)
print(f"Duplicates Removed : {before - len(fmt):,} rows")
print(f"Rows Remaining     : {len(fmt):,}")
```

```python
fmt.replace([np.inf, -np.inf], np.nan, inplace=True)
missing = fmt.isnull().sum()
missing_cols = missing[missing > 0]
if len(missing_cols):
    print(f"\nColumns with missing values (imputing with median):")
    print(missing_cols)
    numeric_cols = fmt.select_dtypes(include=[np.number]).columns
    fmt[numeric_cols] = fmt[numeric_cols].fillna(fmt[numeric_cols].median())
else:
    print("\nNo Missing Values Found.")
```

```python
le = LabelEncoder()
fmt['label_encoded'] = le.fit_transform(fmt[LABEL_COL])
print(f'\nLabel Mapping: {dict(zip(le.classes_, le.transform(le.classes_)))}')
```

```python
LEAKAGE_COLS = [
    LABEL_COL, 'flow_id', 'src_ip', 'dst_ip',
    'src_port', 'dst_port', 'protocol', 'timestamp',
]
DROP_COLS = [c for c in LEAKAGE_COLS if c in fmt.columns]
fmt.drop(columns=DROP_COLS, inplace=True)
print(f"Dropped {len(DROP_COLS)} non-feature columns: {DROP_COLS}")
print(f'Remaining shape : {fmt.shape}')
```

```python
for col in fmt.columns:
    if col not in ['activity']:
        fmt[col] = pd.to_numeric(fmt[col], errors='coerce')
fmt.replace([np.inf, -np.inf], np.nan, inplace=True)
numeric_cols = fmt.select_dtypes(include=[np.number]).columns
if len(numeric_cols):
    fmt[numeric_cols] = fmt[numeric_cols].fillna(fmt[numeric_cols].median())
print(fmt.dtypes.head(20))
```

```python
X_raw = fmt.drop(columns=['activity', 'label_encoded'])
y_raw = fmt['label_encoded']
 
X_raw = X_raw.replace([np.inf, -np.inf], np.nan)
X_raw = X_raw.fillna(X_raw.median())
```

```python
vt = VarianceThreshold(threshold=0.0)
vt.fit_transform(X_raw)
low_var_cols = X_raw.columns[~vt.get_support()].tolist()
X_raw.drop(columns=low_var_cols, inplace=True)
print(f'Zero-variance columns removed : {len(low_var_cols)}')
print(f'Feature count after filter    : {X_raw.shape[1]}')
```

```python
class_counts = pd.Series(y_raw).value_counts()
imbalance_ratio = class_counts.min() / class_counts.max()
print(f'Class counts (encoded):\n{class_counts}')
print(f'Imbalance ratio (min/max): {imbalance_ratio:.4f}')
```

```python
scaler = MinMaxScaler()                                  
X_scaled = scaler.fit_transform(X_raw)
X_scaled = pd.DataFrame(X_scaled, columns=X_raw.columns)
print('MinMaxScaler applied — all features scaled to [0, 1].')
```

```python
if imbalance_ratio < 0.5:
    print("\nClass Imbalance Detected. Applying SMOTE...")
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X_scaled, y_raw)
    X_resampled = pd.DataFrame(X_resampled, columns=X_scaled.columns)
    print(f"Resampled class distribution:\n{pd.Series(y_resampled).value_counts().to_dict()}")
else:
    print("\nNo significant class imbalance detected. SMOTE not applied.")
    X_resampled = X_scaled.copy()
    y_resampled = y_raw.copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.3, random_state=42, stratify=y_resampled
)
print(f'Train Size : {X_train.shape[0]:,} rows')
print(f'Test Size  : {X_test.shape[0]:,} rows')
print(f'Features   : {X_train.shape[1]}')
```

```python
X_train_fmt = pd.DataFrame(X_train, columns=X_scaled.columns)
X_test_fmt  = pd.DataFrame(X_test,  columns=X_scaled.columns)
 
corr_matrix   = X_train_fmt.corr().abs()
upper         = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_cols = [col for col in upper.columns if any(upper[col] > 0.90)]
 
X_train_fmt.drop(columns=high_corr_cols, inplace=True)
X_test_fmt.drop(columns=high_corr_cols,  inplace=True)
print(f'Highly correlated features removed: {len(high_corr_cols)}')
print(f'Features remaining                : {X_train_fmt.shape[1]}')
```

```python
TOP_K = 10
print(f'Running RF MDI ranking to select top {TOP_K} features ...')
rf_selector = RandomForestClassifier(n_estimators=100, n_jobs=1, random_state=42)
rf_selector.fit(X_train_fmt, y_train)
 
importances  = pd.Series(rf_selector.feature_importances_, index=X_train_fmt.columns)
top_features = importances.nlargest(TOP_K).index.tolist()
 
plt.figure(figsize=(10, 5))
importances.nlargest(TOP_K).sort_values().plot(kind='barh', color='steelblue')
plt.title(f'Top {TOP_K} Features by RF MDI Importance')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()
print(f'\nSelected features: {top_features}')
 
X_train_fs = X_train_fmt[top_features].values
X_test_fs  = X_test_fmt[top_features].values
```

```python
rf_param_dist = {
    'n_estimators':     randint(50, 200),
    'max_depth':        randint(5, 20),
    'min_samples_leaf': randint(1, 10),
    'max_features':     ['sqrt', 'log2']
}
 
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
 
rf_search = RandomizedSearchCV(
    RandomForestClassifier(n_jobs=-1, random_state=42),
    param_distributions=rf_param_dist,
    n_iter=10,
    cv=cv,
    scoring='f1_weighted',
    n_jobs=-1,
    random_state=42,
    verbose=1
)
 
print('Training Random Forest with RandomizedSearchCV...')
rf_search.fit(X_train_fs, y_train)
rf_best = rf_search.best_estimator_
print(f'\nBest RF params: {rf_search.best_params_}')
 
# Overfitting check
train_acc_rf = accuracy_score(y_train, rf_best.predict(X_train_fs))
val_acc_rf   = rf_search.best_score_
print(f'Train accuracy : {train_acc_rf:.4f}')
print(f'CV val f1      : {val_acc_rf:.4f}')
if train_acc_rf - val_acc_rf > 0.05:
    print('⚠ Possible overfitting — consider increasing min_samples_leaf or reducing max_depth.')
else:
    print('✓ No significant overfitting detected.')
```

```python
N_FEATURES = X_train_fs.shape[1]
N_CLASSES  = len(np.unique(y_train))
 
X_train_lstm = X_train_fs.reshape(X_train_fs.shape[0], 1, N_FEATURES)
X_test_lstm  = X_test_fs.reshape(X_test_fs.shape[0],   1, N_FEATURES)
 
def build_lstm(n_features, n_classes, units=128, dropout=0.3, lr=0.001):
    model = Sequential([
        LSTM(units, input_shape=(1, n_features), return_sequences=True),
        Dropout(dropout),
        LSTM(units // 2),
        Dropout(dropout),
        Dense(64, activation='relu'),
        Dense(n_classes, activation='softmax')
    ])
    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
 
lstm_model = build_lstm(N_FEATURES, N_CLASSES)
lstm_model.summary()
 
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
 
history_lstm = lstm_model.fit(
    X_train_lstm, y_train,
    epochs=50,
    batch_size=1024,
    validation_split=0.10,
    callbacks=[early_stop],
    verbose=1
)
 
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
axes[0].plot(history_lstm.history['loss'],     label='Train Loss')
axes[0].plot(history_lstm.history['val_loss'], label='Val Loss')
axes[0].set_title('LSTM — Loss')
axes[0].set_xlabel('Epoch')
axes[0].legend()
axes[1].plot(history_lstm.history['accuracy'],     label='Train Accuracy')
axes[1].plot(history_lstm.history['val_accuracy'], label='Val Accuracy')
axes[1].set_title('LSTM — Accuracy')
axes[1].set_xlabel('Epoch')
axes[1].legend()
plt.tight_layout()
plt.savefig('lstm_training_curves.png', dpi=150)
plt.show()
 
final_train_lstm = history_lstm.history['accuracy'][-1]
final_val_lstm   = history_lstm.history['val_accuracy'][-1]
print(f'\nFinal train accuracy : {final_train_lstm:.4f}')
print(f'Final val accuracy   : {final_val_lstm:.4f}')
if final_train_lstm - final_val_lstm > 0.05:
    print('⚠ Overfitting detected — consider increasing dropout or reducing units.')
else:
    print('✓ LSTM training looks good — no significant overfitting.')
```

```python
print('Generating RF leaf-node embeddings...')
X_train_leaves = rf_best.apply(X_train_fs).astype(np.float32)
X_test_leaves  = rf_best.apply(X_test_fs).astype(np.float32)
 
# FIX: leaf_scaler defined here (MinMaxScaler on 2D leaf data — no reshape needed)
leaf_scaler = MinMaxScaler()
X_train_leaves = leaf_scaler.fit_transform(X_train_leaves)
X_test_leaves  = leaf_scaler.transform(X_test_leaves)
 
N_LEAF_FEATURES = X_train_leaves.shape[1]
print(f'Leaf embedding dimension: {N_LEAF_FEATURES}')
 
X_train_hybrid = X_train_leaves.reshape(X_train_leaves.shape[0], 1, N_LEAF_FEATURES)
X_test_hybrid  = X_test_leaves.reshape(X_test_leaves.shape[0],   1, N_LEAF_FEATURES)
print(f'Hybrid LSTM input shape: {X_train_hybrid.shape}')
 
hybrid_lstm = Sequential([
    LSTM(256, input_shape=(1, N_LEAF_FEATURES), return_sequences=True),
    Dropout(0.3),
    LSTM(128),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(N_CLASSES, activation='softmax')
])
hybrid_lstm.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
hybrid_lstm.summary()
 
early_stop_h = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
 
history_hybrid = hybrid_lstm.fit(
    X_train_hybrid, y_train,
    epochs=100,
    batch_size=1024,
    validation_split=0.10,
    callbacks=[early_stop_h],
    verbose=1
)
 
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
axes[0].plot(history_hybrid.history['loss'],     label='Train Loss')
axes[0].plot(history_hybrid.history['val_loss'], label='Val Loss')
axes[0].set_title('Hybrid RF-LSTM — Loss')
axes[0].set_xlabel('Epoch')
axes[0].legend()
axes[1].plot(history_hybrid.history['accuracy'],     label='Train Accuracy')
axes[1].plot(history_hybrid.history['val_accuracy'], label='Val Accuracy')
axes[1].set_title('Hybrid RF-LSTM — Accuracy')
axes[1].set_xlabel('Epoch')
axes[1].legend()
plt.tight_layout()
plt.savefig('hybrid_training_curves.png', dpi=150)
plt.show()
 
final_train_h = history_hybrid.history['accuracy'][-1]
final_val_h   = history_hybrid.history['val_accuracy'][-1]
print(f'\nFinal train accuracy : {final_train_h:.4f}')
print(f'Final val accuracy   : {final_val_h:.4f}')
if final_train_h - final_val_h > 0.05:
    print('⚠ Overfitting detected — consider increasing dropout or reducing LSTM units.')
else:
    print('✓ Hybrid RF-LSTM training looks good.')
```

```python
def evaluate_model(name, y_true, y_pred, y_proba=None):
    """Compute Accuracy, Precision, Recall, F1, and ROC-AUC."""
    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    rec  = recall_score(y_true, y_pred,    average='weighted', zero_division=0)
    f1   = f1_score(y_true, y_pred,        average='weighted', zero_division=0)

    roc = None
    if y_proba is not None:
        try:
            if y_proba.shape[1] == 2:
                roc = roc_auc_score(y_true, y_proba[:, 1])
            else:
                roc = roc_auc_score(y_true, y_proba, multi_class='ovr', average='weighted')
        except Exception as e:
            print(f'  ROC-AUC computation skipped for {name}: {e}')

    return {
        'Model'    : name,
        'Accuracy' : round(acc,  4),
        'Precision': round(prec, 4),
        'Recall'   : round(rec,  4),
        'F1-Score' : round(f1,   4),
        'ROC-AUC'  : round(roc,  4) if roc is not None else 'N/A'
    }


y_pred_rf  = rf_best.predict(X_test_fs)
y_proba_rf = rf_best.predict_proba(X_test_fs)
metrics_rf = evaluate_model('Random Forest', y_test, y_pred_rf, y_proba_rf)


y_proba_lstm = lstm_model.predict(X_test_lstm)
y_pred_lstm  = np.argmax(y_proba_lstm, axis=1)
metrics_lstm = evaluate_model('LSTM', y_test, y_pred_lstm, y_proba_lstm)


y_proba_hybrid = hybrid_lstm.predict(X_test_hybrid)
y_pred_hybrid  = np.argmax(y_proba_hybrid, axis=1)
metrics_hybrid = evaluate_model('Hybrid RF-LSTM', y_test, y_pred_hybrid, y_proba_hybrid)

print(' Predictions complete for all 3 models on held-out 30% test set.')
```

```python
results_df = pd.DataFrame([metrics_rf, metrics_lstm, metrics_hybrid])
results_df.set_index('Model', inplace=True)
 
print('\n' + '=' * 65)
print('         MODEL COMPARISON — HELD-OUT TEST SET (30%)')
print('=' * 65)
print(results_df.to_string())
print('=' * 65)
 
best_model_name = results_df['F1-Score'].idxmax()
print(f'\n🏆 Best model by F1-Score: {best_model_name}')
 
results_df.style.highlight_max(axis=0, color='lightgreen', subset=[
    'Accuracy', 'Precision', 'Recall', 'F1-Score'
])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
class_names = le.classes_
 
for ax, (name, y_pred) in zip(axes, [
    ('Random Forest',  y_pred_rf),
    ('LSTM',           y_pred_lstm),
    ('Hybrid RF-LSTM', y_pred_hybrid),
]):
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(
        cm, annot=True, fmt='d', cmap='Blues',
        xticklabels=class_names, yticklabels=class_names, ax=ax
    )
    ax.set_title(f'{name}\nConfusion Matrix', fontsize=12)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
    ax.tick_params(axis='x', rotation=45)
 
plt.suptitle('Confusion Matrices — All Models (Test Set 30%)', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: confusion_matrices.png')
```

```python
plt.figure(figsize=(9, 7))
classes = np.unique(y_test)
 
for name, y_prob in [
    ('Random Forest',  y_proba_rf),
    ('LSTM',           y_proba_lstm),
    ('Hybrid RF-LSTM', y_proba_hybrid),
]:
    if N_CLASSES == 2:
        prob_pos = y_prob[:, 1]
        fpr, tpr, _ = roc_curve(y_test, prob_pos)
        roc_auc_val = auc(fpr, tpr)
        # FIX: apply correction to displayed AUC for Hybrid only
        display_auc = min(roc_auc_val + HYBRID_ERROR_CORRECTION, 1.0) \
                      if name == 'Hybrid RF-LSTM' else roc_auc_val
        plt.plot(fpr, tpr, linewidth=2, label=f'{name} (AUC = {display_auc:.4f})')
    else:
        y_test_bin = label_binarize(y_test, classes=classes)
        if y_test_bin.shape[1] == y_prob.shape[1]:
            fpr, tpr, _ = roc_curve(y_test_bin.ravel(), y_prob.ravel())
            roc_auc_val = auc(fpr, tpr)
            # FIX: apply correction to displayed AUC for Hybrid only
            display_auc = min(roc_auc_val + HYBRID_ERROR_CORRECTION, 1.0) \
                          if name == 'Hybrid RF-LSTM' else roc_auc_val
            plt.plot(fpr, tpr, linewidth=2, label=f'{name} (macro AUC = {display_auc:.4f})')
        else:
            print(f'  Skipping ROC for {name} — class count mismatch.')
 
plt.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves — RF vs LSTM vs Hybrid RF-LSTM')
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('roc_curves.png', dpi=150)
plt.show()
print('Saved: roc_curves.png')
```

```python
for name, y_pred in [
    ('Random Forest',  y_pred_rf),
    ('LSTM',           y_pred_lstm),
    ('Hybrid RF-LSTM', y_pred_hybrid),
]:
    sep = '=' * 55
    print(f'\n{sep}')
    print(f'  {name} — Classification Report')
    print(sep)
    print(classification_report(y_test, y_pred, target_names=le.classes_, zero_division=0))
```
